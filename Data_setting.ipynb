```python
# ==============================
# UNZIP + SPLIT YOLO DATASET
# ==============================

import os
import random
import shutil
import zipfile

# ------------------------------
# PATHS
# ------------------------------

zip_path = "/content/drive/MyDrive/ViharaTech/YOLO_Project/Final_DataSet.zip"   # uploaded zip file
extract_path = "/content/dataset"

# ------------------------------
# UNZIP DATASET
# ------------------------------

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Dataset Extracted Successfully!")
extract_path = "/content/dataset/Car_Images"

# ------------------------------
# CREATE YOLO FOLDERS
# ------------------------------

base_dir = "/content/yolo_dataset"

folders = [
    "images/train",
    "images/val",
    "labels/train",
    "labels/val"
]

for folder in folders:
    os.makedirs(os.path.join(base_dir, folder), exist_ok=True)

# ------------------------------
# GET IMAGE FILES
# ------------------------------

image_extensions = [".jpg", ".jpeg", ".png"]

all_images = []

for file in os.listdir(extract_path):
    if os.path.splitext(file)[1].lower() in image_extensions:
        all_images.append(file)

print(f"Total Images Found: {len(all_images)}")

# ------------------------------
# SHUFFLE & SPLIT
# ------------------------------

random.shuffle(all_images)

split_ratio = 0.8   # 80% train, 20% val

split_index = int(len(all_images) * split_ratio)

train_images = all_images[:split_index]
val_images = all_images[split_index:]

# ------------------------------
# COPY FILES
# ------------------------------

def copy_files(image_list, split_type):

    for image_file in image_list:

        image_name = os.path.splitext(image_file)[0]

        label_file = image_name + ".txt"

        # source paths
        image_src = os.path.join(extract_path, image_file)
        label_src = os.path.join(extract_path, label_file)

        # destination paths
        image_dst = os.path.join(base_dir, f"images/{split_type}", image_file)
        label_dst = os.path.join(base_dir, f"labels/{split_type}", label_file)

        # copy image
        shutil.copy(image_src, image_dst)

        # copy label if exists
        if os.path.exists(label_src):
            shutil.copy(label_src, label_dst)

# copy train and val data
copy_files(train_images, "train")
copy_files(val_images, "val")

# ------------------------------
# DONE
# ------------------------------

print("Dataset Split Completed!")

print(f"Train Images: {len(train_images)}")
print(f"Validation Images: {len(val_images)}")
```
